```python
#Importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading csv file containing diabetes data
diab_data = pd.read_csv('diabetes_model.csv')
```

```python
#Viewing records
diab_data.head()
```

# 1. Create a Count plot of Outcome

```python
sns.countplot(x ='Outcome', data = diab_data)
plt.title('Count plot of Outcome',fontsize=15)
plt.show()
```

# 2. Analyse the effect of age on Outcome using Boxplot

```python
plt.figure(figsize = (10,5) )
sns.boxplot(x= 'Outcome', y='Age',data = diab_data)
plt.title('Box plot of effect of Age on Outcome',fontsize=16)
plt.xlabel('Outcome')
plt.ylabel('Age')
plt.show()
```

From the boxplot it is clear that females belonging to the age group 30-40 are having higher chances of getting diagnosed with diabetes.

# 3. Check for the chances of getting diabetes if it is common in family, using the Diabetes Pedigree Function

```python
plt.figure(figsize = (8,8))
sns.stripplot(x=diab_data['Outcome'],y= diab_data['DiabetesPedigreeFunction'],data = diab_data)
plt.title('Chances of getting Diabetes',fontsize=16)
plt.ylabel('DiabetesPedigreeFunction')
plt.xlabel('Outcome')
plt.show()
```

As the DiabetesPedigreeFunction increases, the chances of getting diabetes are higher. But it depends on the other factors too.

# 4. Check for the relationship with Insulin & Glucose and group the datapoints by Outcome.

```python
plt.figure(figsize = (20,5))
plt.title('Relationship with Insulin & Glucose by Outcome',fontsize=16)
sns.scatterplot(x= 'Insulin', y='Glucose',hue = 'Outcome',  data = diab_data)
plt.show()
```

# 5. Create a heatmap on Correlation of features in the data.

```python
plt.figure(figsize = (8,8) )
plt.title('Heatmap on Correlation of features in the data',fontsize=16,pad=20)
sns.heatmap(diab_data.corr(),annot = True)
plt.xticks(rotation = 90)
plt.show()
```

From the above heatmap, it is clear that all the 8 features are having postive impact on the Outcome. It can be concluded that the features Glucose, BMI, Age, and Pregnacies highly influence the Outcome. That is chances of getting diagnosed with diabetics increases/decreases with an increase/decrease in the values of these features. Of these features Glucose is having higher impact on the Outcome.
